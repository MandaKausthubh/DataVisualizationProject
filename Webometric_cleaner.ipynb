```python
import pandas as pd
```

```python
csv_name = "Webometric_2012.csv"
```

```python
df = pd.read_csv(csv_name)
df
```

```python
year = csv_name[11:15]
df["Year"] = year
```

```python
'''
Eliminate zero values for ranks. They send them to top ranks and will mess with results.
'''

impactZero = df["Impact Rank"] == 0
opennessZero = df["Openness Rank"] == 0
excellenceZero = df["Excellence Rank"] == 0

df.drop(df[impactZero | opennessZero | excellenceZero].index)
df
```

```python
'''
There were some Country with NaNs
'''

df[df.isna()["Country"]]
```

```python
df = df.drop(df[df.isna()["Country"]].index)
```

```python
df["University"] = df["University"].str.replace(r"\(\d+\)\s*", "", regex=True)
```

```python
df
```

```python
df.to_csv(f"cleaned/{csv_name}", index=False)
```
